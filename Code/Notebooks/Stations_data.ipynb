```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

stations_df = pd.read_csv('/content/drive/MyDrive/CS249-Winter2024-Project/california_stations.csv', delimiter=',')
stations_df.iloc[:3]
```

```python
import folium

# Create a map centered at the average location of all stations
map_center = [stations_df['LATITUDE'].mean(), stations_df['LONGITUDE'].mean()]
stations_map = folium.Map(location=map_center, zoom_start=6)

# Add markers for each station
for idx, row in stations_df.iterrows():
    marker = folium.Marker(location=[row['LATITUDE'], row['LONGITUDE']],
                           popup=f"{row['NAME']}, Elevation: {row['ELEVATION']}m")
    marker.add_to(stations_map)

# Display the map
stations_map
```
